### Exercise 2 [Matplotlib's Object Oriented API]
---

Matplotlib have an object-oriented API used to create more complex figures. Code is a
little more complicated, but the advantage is that we now have full control of where the plot axes.

#### Create the same chart as 1.1 using the new method: instantiate figure objects and then call methods or attributes from that object.

```python
x=np.linspace(1,5,10)
y=np.square(x)

fig, ax = plt.subplots()

ax.plot(x,y,c='red')
ax.set_xlabel("X")
ax.set_ylabel("Y = X**2")
ax.set_title("Plot")
```

#### 1) Create the figure below (the blue curve represents $y=x^2$ while the red curve represents $x=y^2$)

```python
# Creating a figure and a main axis object again
fig, main_ax = plt.subplots()

# Ensure x and y have the same length
x=np.linspace(0,2.2,10)
y=np.square(x)


# Plotting x = y^2 (which is y = sqrt(x) if we solve for y) in red on the main axis
main_ax.plot(y**2, y, 'r-', marker='o')
main_ax.set_title('Global Title')
main_ax.legend()
main_ax.set_xlabel("X label original") 
main_ax.set_ylabel("Y label original") 
main_ax.set_xlim(0, 25)
main_ax.set_ylim(0.0, 5.0)

# Creating an inset axis within the main plot for the blue curve y = x^2
# Positioning the inset at the bottom right of the main plot
inset_ax = main_ax.inset_axes([0.5, 0.15, 0.4, 0.4])  # x0, y0, width, height in percentages of the parent's bbox
inset_ax.plot(x, y, 'b-')
inset_ax.set_title('Nested Title')
inset_ax.legend()
inset_ax.set_xlabel("X label nested") 
inset_ax.set_ylabel("Y label nested")
inset_ax.set_xlim(0.0, 1.5)
inset_ax.set_ylim(0.0, 2.0)

# Displaying the plot
plt.show()
```

#### 6) Plot the first 200 rows of the dataframe.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Création du DataFrame
df = pd.DataFrame(np.random.rand(5000, 4)*3 , columns=['col_1', 'col_2', 'col_3', 'col_4'])

# Sélection des 200 premières lignes
subset_df = df.head(200)

# Plot avec remplissage entre la ligne et l'axe des x
ax = subset_df.plot(figsize=(10, 4))
ax.set_ylim(0.0, 3.5)  # Définition de la plage de l'axe des y de 0.0 à 3.5

# Couleurs de remplissage pour chaque colonne
fill_colors = ['green', 'red', 'blue', 'orange']

# Remplissage de la zone sous les lignes pour chaque colonne
for column, color in zip(subset_df.columns, fill_colors):
    ax.fill_between(range(200), subset_df[column], 0, where=(subset_df[column] > 0), facecolor=color, alpha=0.3, interpolate=True)

plt.show()
```

#### 7) Create two plots side by side, the first one is using scatter plot (as seen previously) and the second one use hexbin plot.

```python
fig, axes = plt.subplots(nrows=1,ncols=2,figsize=(20,8),squeeze=False)

df.plot.scatter('col_1','col_2',color='blue',ax=axes[0,0])
df.plot.hexbin('col_1','col_2',gridsize=20,cmap='magma',ax=axes[0,1])
```
